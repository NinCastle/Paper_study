# 뉴스와 주가 : 빅데이터 감성분석을 통한 지능형 투자의사결정 모형

위 논문은 감성분석을 통하여 주가에 미치는 영향도를 파악한다.

분석방법: 오피니언 마이닝 -> 주가지수의 등락을 예측 -> 지능형 투자의사결정 모형 제시  
  
분석결과: 뉴스 컨텐츠의 감성분석과 주가지수의 등락과는 유의미한 관계를 가지고 있다.  

## 뉴스 컨텐츠와 주가지수
* 주식시장 개장 전 뉴스와 주가지수의 등락과의 관계는 통계적으로 유의미.  
  
## 유의미한 뉴스
> 시황
> 전망
> 해외 뉴스

위의 3가지 뉴스가 가장 변동을 잘 예측하는 것으로 판단됨.  
> 로지스틱 회귀분석 결과 분류정확도는 
* 주가하락시 70%
* 주가상승시 78.8%
* 전체 평균 74.6%

## 1. 서론

뉴스와 주가는 너무다도 다양한 요인들이 존재한다.  
그리고 하루에도 너무 많은 뉴스가 등장하므로 주가에 미치는 영향을 파악하기 거의 불가능하다.  
또한 뉴스의 특성상 중립적 뉘양스가 많고 주식시장의 긍부정을 동시에 제시하기에 긍정과 부정을 나눠 파악하기 어렵다.  
또한 뉴스를 분석하는 주체에 따라 주관이 달라질 위험성도 있다.  
때문에 특성 뉴스를 지정하고 오피니언 마이닝 기법 이용하여 비정형 뉴스의 컨텐츠 감성을 파악하고 뉴스가 주식시장에 미칠 영향을 분석하였으며,  
이를 통해 투자정보를 생성하는 투자의사결정 모형을 제시한다.

## 2. 관련 연구
전통적 연구의 관점은 펀더멘털 관점에서 시작하였으나 만족할만한 예측결과를 얻지못 하였다.  
따라서 인공지능을 이용한 주가예측이 시도되었다.  
> 기술적 지표위주의 주가 분석 및 예측

* 김선웅, 안현철(2010) 기술적 지표외 주가에 영향을 미치는 다양한 비가격 변수를 추가하고 주가예측의 성과를 누적 투자수익률 관점에서 분석
* 김진화 등(2011)은 지식 누석을 이용한 실시간 주식시장 예측 연구를 통해 주가가 연속발생 데이터라는 점에서 마이닝 대상 자료의 변화를 다루는 점진적 데이터마이닝 분석시도
  
> 뉴스와 주가간의 관계를 분석  
>(해외)

* Veronsei(1999) 호황기 부정적인 뉴스에 주식시장이 반응하는 정도에 대한 연구를 실시
* Lee Gillam(2002) 뉴스의 Good/Bad 빈도를 측정하여 FTSE 100 지수(Financial Times Stock Exchange : 런던 증권 거래소에 상장되어 있는 주식 중, 시가 총액 순서대로 100개 기업의 주가를 지수화환 종합 주가 지수)와 비교
* Tak-chung Fu(2008) 홍콩주식 시장을 대상으로 주가흐름과 뉴스 사이의 핵심연계 지점을 도출  
  
>(국내)  
* 송치영(2002) 주가의 일변 변동폭이 큰 200개의 표본을 추출하여 뉴스와 비교하여, 뉴스의 발생이 주가에 중요한 영향을 미쳤으며 뉴스 유형벌로 유, 불리 뉴스에 주가가 다르게 반응하는 비대칭적 반응을 보이고 있음을 실증 분석
* 이근영(2006)은 전쟁위험의 증대가 주식시장의 주가를 하락시킨다고 파악
* 안희준(2010)은 남북관계에서 일어나는 특정 사건들과 주식시작에 상관관계를 분석한 결과 유의미한 영향이 있는것을 분석
* 신종화, 이의철(2010) 인터넷 매체들이 시장에 매우 큰 영향이 미치고 있음을 여러 사례를 통해 확인
* 안성원, 조성배(2010) 뉴스텍스트에서 추출된 특성을 주가에 변동포과 대비하여 Naive Bayerrsian 분류기로 학습 후 개별 종목 주가의 상승, 하락을 예측하는 모형을 제시

## 3. 지능형 투자의사결정모형
1. 구조
    1. 실시간으로 뉴스를 크롤링하여 뉴스 정보에 인덱싱 적용(뉴스 내용 뿐만 아니라 매체, 시간, 뉴스 융형등 다양한 정보를 수집, 분류하여 투자의사결정을 유추할 수 있는 변수로서 재가공)
    2. 뉴스 컨텐츠의 텍스트를 형태소로 분리하고 긍부정 사정과 비교하여 단어별 긍부정 극성을 태깅한다.
    3. 인덱싱 된 분류정보와 스코어링룰을 이용하여 뉴스의 긍부정 극성을 판별하고 이를 다시 일별 스코어링 기준에 의해 최종 투자의사결정에 정보를 도출한다.
    4. 긍부정 극성을 판별하는 계산하는 식  
        > 1. 1개의 뉴스에 대한 긍정/부정비율 NewsPNr을 계산
        $$
        NewsPNr = \frac{\sum_{i=1}^n wordPN(i)}{n} \times 100%
        $$
            1. 하나의 뉴스에는 여러 단어가 존재하며, 각 단어는 긍정/부정의 극성을 가진다.
            2. 각 단어를 긍정/부정 감성 사전과 비교하여 극성태깅한다.  
            WordPN(i) = 단어의 긍정/부정값(1 or 0)   
            3. 각 단어들의 긍정/부정 값을 뉴스의 긍정/부정 비율로 합산.  
            
        > 2. 뉴스에 대한 긍정/부정 판별 NewsPN(P, N, NEU)조건  
        파이썬 코드로 표현
        ```python
        if NewsPNr > 52:
            NewsPN = 'Positive'
        elif 52 >= NewsPNr > 48:
            NewsPN = 'Neu'
        elif NewsPNr < 48:
            NewsPN = 'Negative'
        ```  
        
        > 3. 하루 동안 발생한 모든 뉴스의 긍정/부정을 도출하는 식
        $$
        DayPNr = \frac{(NewsPNr_1 + NewsPNr_2 + \ldots + NewsPNr_n)}{n} = \frac{\sum_{j=1}^n\;NewsPNr(j)}{n}
        $$
          1.뉴스의 인덱싱 정보와 스코어링 룰에 따라 다양한 변수들로 재가공
          $$Day.X.PNr = \frac{\sum_{j=1}^n News.X.PNr(j)}{n}$$
          2.하루 동안 발생한 전망(forecast) 뉴스들의 긍정/부정 스코어
          $$Day.fcPNr = \frac{\sum_{j=1}^n News.fcPNr(j)}{n}$$  
          
        >   d. 투자의사결정정보는 투자의사결정 회귀함수 $\hat{Y}$를 통해 Buy/Sell 의견을 제시한다.
        $$ \hat{Y} = b_0 + b_1 X_1 + b_2 X_2 + \ldots + b_n X_n$$

        * 뉴스의 유형에 따라 주가에 미치는 영향이 다를 것.

## 4. 실증 분석 결과  

1. 뉴스와 주가의 변동곡선  

주가지수 등락의 흐름과 일자별 뉴스의 긍부정 마이닝 결과 두 그래프가 매우 유사한 흐름을 보이고 있음을 알 수 있다.  
따라서 대체로 긍정의견 일경우 주가가 상승하고 부정의견 일경우 주가가 하락하는 것을 보여준다.  
또한 특이점으로 개장 전 뉴스와 주가의 흐름을 비교했을때 주가의 흐름과 유사함을 보였다.  
위의 유사점을 통하여 뉴스의 긍정/부정 패턴이 매우 밀접한 관계를 가지고 있으며,  
특히 장전뉴스가 주가에 일정수준 선행하고 있거나 영향을 미치고 있다고 볼 수 있다.  

2. 뉴스와 주가 등락  

    * 주식시장 전후 시간별로 구분하여 이때의 긍정/부정 마이닝 결과가 주가의 등락에 유의미한것으로 분석되었다.  
그정 평균에 따르면 주가가 오르던 내리던 관계없이 뉴스의 기조는 대체로 부정적이며 주가 상승시 덜 부정적으로 파악됨.
    * 뉴스 유형별
        * 환율의 변화
        * 경기 흐름
        * 정부 정책
        * 시황 뉴스
        * 전망 뉴스(주가의 상승/하락을 예측)
        * 업종 뉴스(경제 여건에 따라 영향을 받거나 특별하게 언급된 뉴스)
        * 종목 뉴스(개별 기업의 성과)
        * 해외 뉴스(미국, 유럽)
    시황, 전망, 해외뉴스만 유의미
    특이점으로 개별 기업 종목 뉴스와 같은 경우는 하락시장에서의 오피니언 긍정 값이 높게 나타났는데,  
    이는 주가하락을 방어하고자 하는 긍정적 내용의 홍보기획 기사들이 다수 출현했기 때문으로 유추
    * 뉴스를 단순점수로 비교하는 것보다 비율값으로 산출 분석하는것이 더 높은 설명력을 가진다는 것을 확인
    * 시황, 전망, 해외뉴스 변수만을 이용해야 좋은 결과를 도출

## **주가하락**에 대해서 70.0%, **주가상승**에 대해서는 78.8%로 주가하락보다는 주가상승에 대한 분류 정확도가 조금 더 높았으며, **전체적**으론 74.6%의 정확도를 보여주고 있다.

## 한계점 
* 분석데이터의 양이 너무 적었다.
